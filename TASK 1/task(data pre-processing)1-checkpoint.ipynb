```python
null = data.isnull().sum()
print(null[null > 0])
```

```python
data['director']=data['director'].fillna('unknown')
null = data.isnull().sum()
print(null[null > 0])
```

```python
data = data.fillna(data.mode().iloc[0])
```

```python
data.head()
```

```python
data['type']=data['type'].astype('category')
data['title']=data['title'].astype('string')
data['director']=data['director'].astype('string')
data['cast']=data['cast'].astype('string')
data['country']=data['country'].astype('category')
data['rating']=data['rating'].astype('category')
data['duration']=data['duration'].astype('string')
data['listed_in']=data['listed_in'].astype('string')
data['description']=data['description'].astype('string')
```

```python
data.release_year= pd.PeriodIndex(data.release_year,freq="Y")
```

```python
data.date_added=pd.to_datetime(data.date_added,format="mixed")
print(data.date_added)
```
