## HANDELLING MISSING VALUES
#### "cabin" has a lot of null values we don't require for analysis
#### "embarked" is a categorious variable and has only 3 unique caterories we fill the null values with the mode of col vlues = "s"
#### "age" is a continous variable the null values can be filled with the median specificaly w.r.t their gender and pclass


```python
data = data.drop(["Cabin"], axis =1)
```

```python
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
```

```python
def null_age(col):
    Age = col.iloc[0]
    Sex = col.iloc[1]
    Pclass = col.iloc[2]
    
    if pd.isnull(Age):
        if Sex == 'female':
            if Pclass == 1:
                return 32
            else:
                return 28
            
        else:  # female
            if Pclass == 1:
                return 36
            elif Pclass == 2:
                return 29
            else:
                return 28
    else:
        return Age


data['Age'] = data[['Age', 'Sex', 'Pclass']].apply(null_age, axis=1)
```

### Droping other unwanted columns ["PassengerId" ,"Name","Ticket"]


```python
data= data.drop(["PassengerId" ,"Name","Ticket"] , axis = 1)
```
